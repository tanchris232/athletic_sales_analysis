```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv("Resources/athletic_sales_2020.csv")
athletic_sales_2021_df = pd.read_csv("Resources/athletic_sales_2021.csv")
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020_df
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
athletic_sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_20_21_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df])
combined_20_21_df.reset_index(drop=True)
```

```python
# Check if any values are null.
clean_combined_20_21_df = combined_20_21_df.dropna(how="any")
clean_combined_20_21_df.count()
```

```python
# Check the data type of each column
clean_combined_20_21_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
clean_combined_20_21_df['invoice_date'] = pd.to_datetime(clean_combined_20_21_df['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
clean_combined_20_21_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
tot_sold_group = clean_combined_20_21_df.groupby(["region", "state", "city"])[["units_sold"]].sum()
tot_sold_group = tot_sold_group.rename(columns={"units_sold": "Total_Products_Sold"})
tot_sold_group

# Show the top 5 results.
tot_sold_group.sort_values(by=["Total_Products_Sold"], ascending=False).head()                          
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
tot_sold_pivot = pd.pivot_table(clean_combined_20_21_df,
                                index=["region", "state", "city"], 
                                values=["units_sold"], 
                                aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
tot_sold_pivot = tot_sold_pivot.rename(columns={"units_sold": "Total_Products_Sold"})

# Show the top 5 results.
tot_sold_pivot.sort_values(by=["Total_Products_Sold"], ascending=False).head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
tot_sales_group = clean_combined_20_21_df.groupby(["region", "state", "city"])[["total_sales"]].sum()
tot_sales_group = tot_sales_group.rename(columns={"total_sales": "Total Sales"})
tot_sales_group

# Show the top 5 results.
tot_sales_group.sort_values(by=["Total Sales"], ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
tot_sales_pivot = pd.pivot_table(clean_combined_20_21_df,
                                index=["region", "state", "city"], 
                                values=["total_sales"], 
                                aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
tot_sales_pivot = tot_sales_pivot.rename(columns={"total_sales": "Total Sales"})

# Show the top 5 results.
tot_sales_pivot.sort_values(by=["Total Sales"], ascending=False).head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
tot_sales_retailer_group = clean_combined_20_21_df.groupby(["retailer", "region", "state", "city"])[["total_sales"]].sum()
tot_sales_retailer_group = tot_sales_retailer_group.rename(columns={"total_sales": "Total Sales"})

# Show the top 5 results.
tot_sales_retailer_group.sort_values(by=["Total Sales"], ascending=False).head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
clean_combined_20_21_df['product'].unique()
```

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_foot_combined_20_21_df = clean_combined_20_21_df[clean_combined_20_21_df['product'] == "Women's Athletic Footwear"]
womens_foot_combined_20_21_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_foot_sold_group = womens_foot_combined_20_21_df.groupby(["retailer", "region", "state", "city"])[["units_sold"]].sum()
womens_foot_sold_group = womens_foot_sold_group.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})
womens_foot_sold_group

# Show the top 5 results.
womens_foot_sold_group.sort_values(by="Womens_Footwear_Units_Sold", ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_foot_sold_pivot = womens_foot_combined_20_21_df.pivot_table(index=["retailer", "region", "state", "city"], 
                                                               values="units_sold",
                                                               aggfunc='sum')
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_foot_sold_pivot = womens_foot_sold_pivot.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})
# Show the top 5 results.
womens_foot_sold_pivot.sort_values(by="Womens_Footwear_Units_Sold", ascending=False).head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
wom_inv_sales = womens_combined_20_21_df.pivot_table(index="invoice_date",
                                                    values="total_sales").round(2)

# Optional: Rename the "total_sales" column to "Total Sales"
wom_inv_sales = wom_inv_sales.rename(columns={"total_sales": "Total Sales"})

# Show the table.
wom_inv_sales.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
wom_inv_sales_daily = wom_inv_sales.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
wom_inv_sales_daily.sort_values(by="Total Sales", ascending=False).head()
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
wom_inv_sales_weekly = wom_inv_sales.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
wom_inv_sales_weekly.sort_values(by="Total Sales", ascending=False).head()
```
